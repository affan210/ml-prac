```python
# Select categorical columns
# categorical_columns = df.select_dtypes(include=['object']).columns

obj_col = []
numeric_col = []
date_col = []
for col in df_merge.columns:
    if df_merge[col].dtype == 'object':
        obj_col.append(col)
    if df_merge[col].dtype == 'float64' or df_merge[col].dtype == 'int64':
        numeric_col.append(col)
    if df_merge[col].dtype == 'datetime64[ns]':
        date_col.append(col)

print(' object:',obj_col,'\n numeric:', numeric_col,'\n date:', date_col)
```
