```python
data = []
for sent in sent_tokenize(corpus):
    words = word_tokenize(sent.lower())
    lem_words = [WordNetLemmatizer().lemmatize(word) for word in words if word not in stopwords.words('english')]
    data.append(' '.join(lem_words))
print(data)
# sentences = sent_tokenize(corpus)
# for i in range(len(sentences)):
#     words = word_tokenize(sentences[i].lower())
#     lem_words = [WordNetLemmatizer().lemmatize(word) for word in words if word not in stopwords.words('english')]
#     sentences[i] = ' '.join(lem_words)
# print(sentences)
```

```python
from nltk import pos_tag

pos_data = [pos_tag(word.split()) for word in data]
pos_data
```
